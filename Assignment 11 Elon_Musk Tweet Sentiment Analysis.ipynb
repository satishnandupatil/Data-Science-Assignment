# Text Processing

```python
elon_text.Text=[Text.strip() for Text in elon_text.Text] # remove both the leading and the trailing characters
elon_text.Text=[Text for Text in elon_text.Text if Text] # removes empty strings, because they are considered in Python as False
elon_text.Text[0:10]
```

### Converting sentense in to lower case

```python
elon_text.Text=elon_text.Text.apply(lambda x: x.lower())
```

### Removing twitter username handlers

```python
elon_text.Text=elon_text.Text.apply(lambda x: re.sub('@[^\s*]+'," ",x)) 
```

### Removing links

```python
elon_text.Text=elon_text.Text=elon_text.Text.apply(lambda x: re.sub('https\S*|<\S*>|\d|mph|am|pm|\Samp|[^a-z]'," ",x)) 
```

```python
elon_text.Text=elon_text.Text.apply(lambda x: re.sub(' \S '," ",x)) 
```

### Splitting the words

```python
elon_text.Text=elon_text.Text.apply(lambda x: x.split())
```

```python
# To identify the sentiment of text
from textblob import TextBlob

from textblob.np_extractors import ConllExtractor

from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.stem import WordNetLemmatizer
from nltk.stem.porter import *
from nltk.classify import NaiveBayesClassifier
from wordcloud import WordCloud
import nltk
nltk.download('vader_lexicon')
import matplotlib.pyplot as plt

```

## Fetching Sentiments 


### a.Using NLTK's Sentimental Analyser
### b.Using TextBlob

```python
# 1 way
def fetch_sentiment_using_SIA(text):
    sid = SentimentIntensityAnalyzer()
    polarity_scores = sid.polarity_scores(text)
    return 'neg' if polarity_scores['neg'] > polarity_scores['pos'] else 'pos'

# 2 way
def fetch_sentiment_using_textblob(text):
    analysis = TextBlob(text)
    return 'pos' if analysis.sentiment.polarity >= 0 else 'neg'
```

## a. Using NLTK's SentimentIntensityAnalyzer

```python
sentiments_using_SIA = elon_text.tidy.apply(lambda tweet: fetch_sentiment_using_SIA(tweet))
pd.DataFrame(sentiments_using_SIA.value_counts())
```

## b. Using TextBlob¶

```python
sentiments_using_textblob =elon_text.tidy.apply(lambda tweet: fetch_sentiment_using_textblob(tweet))
pd.DataFrame(sentiments_using_textblob.value_counts())
```

## A. Most common words in positive tweets¶


```python
def generate_wordcloud(all_words):
    wordcloud = WordCloud(width=800, height=500, random_state=21, max_font_size=100, relative_scaling=0.5, colormap='Dark2').generate(all_words)

    plt.figure(figsize=(14, 10))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis('off')
    plt.show()
```

```python
all_words = ' '.join([text for text in elon_text['tidy'][elon_text.sentiment == 'pos']])
generate_wordcloud(all_words)
```
